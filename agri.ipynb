```python
import matplotlib as plt
from matplotlib import pyplot as plt
```

```python
# question 1 : TOP 7 RICE PRODUCING STATES

rice_by_state = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum()
top7 = rice_by_state.sort_values(ascending=False).head(7)
top7
```

```python
top7.plot(kind='bar', color='blue')
plt.title('Top 7 Rice Producing States')
plt.xlabel('State')
plt.ylabel('Rice Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# question 2 : TOP 5 WHEAT PROCUDING STATES

wheat_by_state = df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum()
top5 = wheat_by_state.sort_values(ascending=False).head(5)
top5.plot(kind='bar', color='brown')
plt.title('Top 5 Wheat Producing States')
plt.xlabel('State')
plt.ylabel('Wheat Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# WHEAT PRODUCING STATES IN PERCENTAGE

top5_percent = 100 * top5 / top5.sum() 
top5_percent.plot(kind='pie',autopct='%1.1f%%')
plt.title('Wheat producing states in Percentage')
plt.ylabel('') 
plt.tight_layout()
plt.show()
```

```python
# question 3 : TOP 5 OILSEED PRODUCING STATES

oilseed_states = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum()
top_5 = oilseed_states.sort_values(ascending=False).head(5)
top_5.plot(kind='bar', color='olive')
plt.title('Top 5 Oilseed Producing States')
plt.xlabel('State')
plt.ylabel('Oilseeds Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# question 4 : 

sunflower_by_state = df.groupby('State Name')['SUNFLOWER PRODUCTION (1000 tons)'].sum()
top7 = sunflower_by_state.sort_values(ascending=False).head(7)
top7.plot(kind='bar', color='yellow')
plt.title('Top 7 Sunflower Producing States')
plt.xlabel('State')
plt.ylabel('Sunflower Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
#question 5 : INDIA SUGARCANE PRODUCTION FOR LAST 50 YEARS

df_50y = df[(df['Year'] >= 1968) & (df['Year'] <= 2017)]
sugarcane_by_year = df_50y.groupby('Year')['SUGARCANE PRODUCTION (1000 tons)'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.plot(sugarcane_by_year['Year'], sugarcane_by_year['SUGARCANE PRODUCTION (1000 tons)'], marker='o', color='darkgreen')
plt.title("India's Sugarcane Production last 50 years")
plt.xlabel("Year")
plt.ylabel("Sugarcane Production (1000 tons)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# QUESTION 6: Rice Production Vs Wheat Production (Last 50y)

df_50y = df[(df['Year'] >= 1968) & (df['Year'] <= 2017)]
rice_vs_wheat = df_50y.groupby('Year')[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)']].sum().reset_index()
plt.figure(figsize=(10,6))
plt.plot(rice_vs_wheat['Year'], rice_vs_wheat['RICE PRODUCTION (1000 tons)'], label='Rice Production', marker='o', color='goldenrod')
plt.plot(rice_vs_wheat['Year'], rice_vs_wheat['WHEAT PRODUCTION (1000 tons)'], label='Wheat Production', marker='s', color='saddlebrown')
plt.title("Rice vs Wheat Production in India last 50years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# question 7: Rice Production By West Bengal Districts

df_wb = df[df['State Name'] == 'West Bengal']
district_wb = df_wb.groupby('Dist Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)
district_wb.plot(kind='bar', color='seagreen', figsize=(12,6))
plt.title("Rice Production by West Bengal Districts")
plt.xlabel("District")
plt.ylabel("Rice Production (1000 tons)")
plt.tight_layout()
plt.show()
```

```python
# question 8: Top 10 Wheat Production Years From UP (bar chart added optionally)

df_up = df[df['State Name'] == 'Uttar Pradesh']
up_yearly_wheat = df_up.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
top10_years = up_yearly_wheat.sort_values('WHEAT PRODUCTION (1000 tons)', ascending=False).head(10)
print(top10_years)
```

```python
plt.bar(top10_years['Year'].astype(str), top10_years['WHEAT PRODUCTION (1000 tons)'], color='saddlebrown')
plt.title('Top 10 Wheat Production Years in UP')
plt.xlabel('Year')
plt.ylabel('Wheat Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# question 9 : Millet Production (Last 50y)  (line plot added optionally)

df_50y = df[(df['Year'] >= 1968) & (df['Year'] <= 2017)]
df_50y['TOTAL_MILLET_PRODUCTION'] = ( df_50y['PEARL MILLET PRODUCTION (1000 tons)'] + df_50y['FINGER MILLET PRODUCTION (1000 tons)'] )
millet_by_year = df_50y.groupby('Year')['TOTAL_MILLET_PRODUCTION'].sum().reset_index()
print(millet_by_year)
```

```python
plt.figure(figsize=(10,6))
plt.plot(millet_by_year['Year'], millet_by_year.iloc[:,1], marker='o', color='teal')
plt.title("India's Millet Production last 50 years")
plt.xlabel("Year")
plt.ylabel("Millet Production (1000 tons)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# question 10: Sorghum Production (Kharif and Rabi) by Region (Stacked Bar Chart added optionally)

region_sorghum = df.groupby('State Name')[['KHARIF SORGHUM PRODUCTION (1000 tons)', 'RABI SORGHUM PRODUCTION (1000 tons)']].sum()
print(region_sorghum )
region_sorghum.plot(kind='bar', stacked=True, color=['coral', 'gold'], figsize=(12,6))
plt.title('Sorghum Production (Kharif and Rabi) by Region')
plt.xlabel('Region')
plt.ylabel('Sorghum Production (1000 tons)')
plt.legend(['Kharif', 'Rabi'])
plt.tight_layout()
plt.show()
```

```python
# question 11: Top 7 States for Groundnut Production (bar chart added optionally)

groundnut_by_state = df.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum()
top7 = groundnut_by_state.sort_values(ascending=False).head(7)
print(top7)
top7.plot(kind='bar', color='brown')
plt.title('Top 7 States for Groundnut Production')
plt.xlabel('State')
plt.ylabel('Groundnut Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
#question 12: Soybean Production by Top 5 States and Yield Efficiency ( charts added optionally)

soybean_state = df.groupby('State Name').agg({'SOYABEAN PRODUCTION (1000 tons)': 'sum', 'SOYABEAN YIELD (Kg per ha)': 'mean'})
top5 = soybean_state.sort_values('SOYABEAN PRODUCTION (1000 tons)', ascending=False).head(5)
print(top5)
```

```python
fig, ax1 = plt.subplots()

# Bar for production
top5['SOYABEAN PRODUCTION (1000 tons)'].plot(kind='bar', color='black', ax=ax1)
ax1.set_ylabel('Production (1000 tons)')
ax1.set_xlabel('State')
plt.title('Top 5 Soybean Producing States and Yield Efficiency')

# Twin axis for yield
ax2 = ax1.twinx()
top5['SOYABEAN YIELD (Kg per ha)'].plot(kind='line', color='orangered', ax=ax2, marker='o')
ax2.set_ylabel('Yield Efficiency')

plt.tight_layout()
plt.show()
```

```python
# question 13 : Oilseed Production in Major States

oilseed_by_state = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum()
major_states = oilseed_by_state.sort_values(ascending=False).head(5)
print(major_states)
major_states.plot(kind='bar', color='olive')
plt.title('Oilseed Production in Major States')
plt.xlabel('State')
plt.ylabel('Oilseed Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# question 14: Impact of Area Cultivated on Production (Rice, Wheat, Maize)

rice = df.groupby('State Name')[['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)']].sum()
wheat = df.groupby('State Name')[['WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)']].sum()
maize = df.groupby('State Name')[['MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)']].sum()
print (rice)
print (wheat)   # print one by one for crct display
print (maize)
```

```python
plt.figure(figsize=(14,4))   # scatter plot added additionally

# Rice
plt.subplot(1,3,1)
plt.scatter(rice['RICE AREA (1000 ha)'], rice['RICE PRODUCTION (1000 tons)'], color='green')
plt.xlabel('Rice Area (1000 ha)')
plt.ylabel('Rice Production (1000 tons)')
plt.title('Rice: Area cultivated on Production')

# Wheat
plt.subplot(1,3,2)
plt.scatter(wheat['WHEAT AREA (1000 ha)'], wheat['WHEAT PRODUCTION (1000 tons)'], color='orange')
plt.xlabel('Wheat Area (1000 ha)')
plt.ylabel('Wheat Production (1000 tons)')
plt.title('Wheat: Area cultivated on Production')

plt.subplot(1,3,3)
plt.scatter(maize['MAIZE AREA (1000 ha)'], maize['MAIZE PRODUCTION (1000 tons)'], color='gold')
plt.xlabel('Maize Area (1000 ha)')
plt.ylabel('Maize Production (1000 tons)')
plt.title('Maize: Area cultivated on Production')

plt.tight_layout()
plt.show()
```

```python
# question 15 : Rice vs Wheat Yield Across States

yield_by_state = df.groupby('State Name').agg({'RICE YIELD (Kg per ha)': 'mean','WHEAT YIELD (Kg per ha)': 'mean'})
print(yield_by_state)
```

```python
import numpy as np
```

```python
states = yield_by_state.index
rice_yield = yield_by_state['RICE YIELD (Kg per ha)']
wheat_yield = yield_by_state['WHEAT YIELD (Kg per ha)']

x = np.arange(len(states))  # label locations
width = 0.35                # width of the bars

plt.figure(figsize=(14,6))
plt.bar(x - width/2, rice_yield, width, label='Rice Yield', color='mediumseagreen')
plt.bar(x + width/2, wheat_yield, width, label='Wheat Yield', color='goldenrod')

plt.xlabel('State')
plt.ylabel('Yield (Kg per ha)')
plt.title('Rice vs Wheat Yield Across States')
plt.xticks(x, states, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
